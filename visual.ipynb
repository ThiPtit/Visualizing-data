```python
# cau 1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('AB_NYC_2019.csv')
sns.barplot( x ='neighbourhood_group', y ='price', data = data, color='pink')
plt.show()
```

```python
# cau 2
sns.histplot(x ='neighbourhood', data =data, kde=True)
plt.show()
```

```python
# cau 3
sns.lineplot(x ='neighbourhood', y = 'availability_365', data = data, color='black')
plt.show()
```

```python
sns.scatterplot(x = 'latitude', y ='longitude', data = data, hue ='neighbourhood')
plt.show()
```

```python
# To heatmap
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', linewidths=0.2)
# phng thc corr()  tnh ton ma trn tng quan gia tt c cc thuc tnh. Kt qu tr v l mt DataFrame.
# Tham s annot=True c s dng  hin th gi tr tng quan trn tng  ca heatmap
# Tham s cmap c s dng  thit lp bn  mu ca heatmap
# coolwarm l bn  mu. 
# Tham s linewidths c s dng  thit lp  rng ca cc ng vin gia cc .

# Hin th biu 
plt.show()
```
